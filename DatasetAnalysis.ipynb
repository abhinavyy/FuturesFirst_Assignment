```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

### LOAD DATASETS

```python
dataset_Data = pd.read_csv("data/Data 2.csv")
print("Data2 Shape:", dataset_Data.shape)
display(dataset_Data.head())

```

```python
stock_dataset = pd.read_csv("data/StockPrice.csv")
print("StockPrice Shape:", stock_dataset.shape)
display(stock_dataset.head())
```

### DATA CLEANING & TYPE CONVERSION

```python
# Convert Date columns
dataset_Data["Date"] = pd.to_datetime(dataset_Data["Date"])
stock_dataset["Date"] = pd.to_datetime(stock_dataset["Date"])

# Sort by Date (IMPORTANT)
dataset_Data = dataset_Data.sort_values("Date").reset_index(drop=True)
stock_dataset = stock_dataset.sort_values("Date").reset_index(drop=True)

print("Data Date Range:", dataset_Data["Date"].min(), "to", dataset_Data["Date"].max())
print("Stock Date Range:", stock_dataset["Date"].min(), "to", stock_dataset["Date"].max())

print("\nMissing values:")
print(dataset_Data.isna().sum())
print(stock_dataset.isna().sum())
print("\nDuplicate Dates:")
print("Data:", dataset_Data["Date"].duplicated().sum())
print("Stock:", stock_dataset["Date"].duplicated().sum())
```

### DESCRIPTIVE STATISTICS

```python
print("DATA STATISTICS")
display(dataset_Data["Data"].describe())

print("STOCK PRICE STATISTICS")
display(stock_dataset["Price"].describe())

print("Skewness:")
print("Data:", dataset_Data["Data"].skew())
print("Price:", stock_dataset["Price"].skew())
```
